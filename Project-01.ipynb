```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import (
    RandomForestRegressor,
    HistGradientBoostingRegressor,
)
import joblib
```

```python
housing = fetch_california_housing() # Carregar dados

x = housing.data # Carregar dados X (input)
y = housing.target # Carregar dados Y (output)

poly = PolynomialFeatures() # Carregamos essa função
x = poly.fit_transform(x) # Essa função aumenta, cria novas possibilidades de dados, isso siginifica mais precisão

# Essa parte do código, estamos separando 20% dos dados para teste
# O resto dos dados, que são 80%, vão ser reservador para treino.
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state = 432  # Essa parte é para focar em um "Dado" em específico, se tirarmos ele, 
                        # ele vai ficar randomizando um número do dataset toda hora
)

# Hyperparameterization

model = HistGradientBoostingRegressor(
    max_iter=350,
    learning_rate=0.05
)
model.fit(x_train, y_train)

joblib.dump(model, "my_model.joblib") # Salvamos esse modelo dessa forma

y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
local_model = joblib.load("my_model.joblib") # Carregamos o modelo que salvamos

y_pred = local_model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print(r2)
```
