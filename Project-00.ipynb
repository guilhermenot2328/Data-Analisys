```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
housing = fetch_california_housing()
x = housing.data
y = housing.target

x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=123)

model = LinearRegression()
model.fit(x_train, y_train) # Essa função ".fit" treina o data set usando o model.
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)

print(r2) # Isso quer dizer que nosso modelo de predição tem 60% de acertos.
```
